```python
import os
import cv2
import numpy as np
import matplotlib.pyplot as plt
import imutils
from google.colab.patches import cv2_imshow
```

```python
def aplicar_sal_pimenta(imagem, tamanho_kernel):

  # Verifica a imagem e se o kernel é ímpar
  if imagem is None:
      print("Erro: Não foi possível carregar a imagem. Verifique o caminho.")
      return
  if tamanho_kernel % 2 == 0:
      print("Erro: O do kernel deve ser um número ímpar (ex: 3, 5, 7).")
      return

  # Filtro da média
  media = cv2.blur(imagem, (tamanho_kernel,tamanho_kernel))


  # Filtro Laplaciano (realce de bordas)
  laplaciano = cv2.Laplacian(imagem, cv2.CV_64F)
  laplaciano = cv2.convertScaleAbs(laplaciano)

  # Realce por subtração (original - média)
  realce = cv2.subtract(imagem, media)

  # Mostrar os resultados
  plt.figure(figsize=(12,6))

  plt.subplot(1,4,1)
  plt.imshow(imagem, cmap='gray')
  plt.title("Original")
  plt.axis('off')

  plt.subplot(1,4,2)
  plt.imshow(media, cmap='gray')
  plt.title(f'Imagem Media (Kernel {tamanho_kernel}x{tamanho_kernel})')
  plt.axis('off')

  plt.subplot(1,4,3)
  plt.imshow(laplaciano, cmap='gray')
  plt.title("Laplaciano")
  plt.axis('off')

  plt.subplot(1,4,4)
  plt.imshow(realce, cmap='gray')
  plt.title("Realce")
  plt.axis('off')

  plt.tight_layout()
  plt.show()


```

```python
# --- Execução do programa ---

caminho_imgagem = '/content/drive/MyDrive/Proc de Imagens 2025 2/Atividade 6/FIG033~4.TIF'

imagem_ori = cv2.imread(caminho_imgagem, cv2.COLOR_BGR2GRAY)

tamanho_kernel = 3

aplicar_sal_pimenta(imagem_ori, tamanho_kernel)
```
